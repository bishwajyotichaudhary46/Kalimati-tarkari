```python
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv("final_data.csv")
```

```python
devanagari_to_english = {
    '०': '0', '१': '1', '२': '2', '३': '3', '४': '4',
    '५': '5', '६': '6', '७': '7', '८': '8', '९': '9'
}

def replace_devanagari_with_eng(text):
    for dev, eng in devanagari_to_english.items():
        text = text.replace(dev, eng)
    return text


```

```python
df_potato= df[df['Commodity'] == 'आलु रातो']
df_potato = df_potato[['Date', 'Average']]
start_date = datetime(2013, 4, 15)
end_date = datetime.now()
df_potato
```

```python
 # Generate the range of dates for the year 2024
date_range = []
current = start_date
end_date = datetime.now()
while current <= end_date:
    date_range.append(current.strftime('%m/%d/%Y'))
    current += timedelta(days=1)

date_range = pd.Index(date_range)
complete_df = pd.DataFrame(date_range, columns=['Date'])
complete_df
```

```python
df_potato['Date'] = pd.to_datetime(df_potato['Date'])
df_potato['Date'] = df_potato['Date'].dt.date
df_potato
```

```python
final_df = pd.merge(complete_df,df_potato, on=['Date'], how='left')
```

```python
df_potato['Average'] = df_potato['Average'].str.replace("रू","")
```

```python
df_potato['Average'] = df_potato['Average'].astype('str').apply(replace_devanagari_with_eng).astype('float')
```
